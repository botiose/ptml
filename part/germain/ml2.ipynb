```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(0, 1, 100)

lambdas = [1, 5, 10]
for l in lambdas:
    plt.plot(x, np.exp(-l*x), label="lambda: " + str(l))
    
plt.xlabel('x')
plt.ylabel('y=exp(-lambda x)')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

fig = plt.figure()

x = np.linspace(0, 1, 100)

lambdas = [1, 5, 10]
for i, l in enumerate(lambdas):
    fig.add_subplot(2, 2, i+1)
    plt.xlabel('x')
    plt.plot(x, np.exp(-l*x), label="lambda: " + str(l))
    plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(0, 1, 100)

lambdas = [1, 5, 10]
for l in lambdas:
    plt.plot(x, np.exp(-l*x), label="lambda: " + str(l))
    plt.annotate("y=1/e", (1/l, np.exp(-1)))
    
plt.xlabel('x')
plt.ylabel('y=exp(-lambda x)')
plt.legend()
plt.show()


```

```python
from matplotlib import pyplot as plt
from matplotlib import patches
import numpy as np
import pandas as pd

data = pd.read_csv('data1.csv')
plt.scatter(data.age, data.fare)
plt.ylabel("tarif")
plt.xlabel("age")
plt.show()

######################

def survived_color(item):
    if (item):
        return 'green'
    return 'yellow'

plt.scatter(data.age, data.fare, c=data.survived.apply(survived_color))
plt.ylabel("tarif")
plt.xlabel("age")

colornames = [patches.Patch(color='green', label='survived'), patches.Patch(color='yellow', label='no')]
plt.legend(handles=colornames)
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import patches
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix


data = pd.read_csv('data1.csv')[['fare', 'age', 'survived', 'sex']].dropna()

colmap = {0:"red", 1:"green"}
colors = list(map(lambda x:colmap.get(x), data.survived))

data.sex = data.sex.apply(lambda x:{'male':0, 'female':1}[x])

scatter_matrix(data[['fare', 'sex', 'age']], alpha=0.5, c=colors, diagonal='hist')
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import patches
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
from random import gauss

data = pd.read_csv('data1.csv')[['fare', 'age', 'survived', 'sex']].dropna()

colmap = {0:"red", 1:"green"}
colors = list(map(lambda x:colmap.get(x), data.survived))

data.sex = data.sex.apply(lambda x:{'male':0, 'female':1}[x])

data.sex = data.sex.apply(lambda x:x+gauss(0, 0.05))

scatter_matrix(data[['fare', 'sex', 'age']], alpha=0.5, c=colors, diagonal='hist')
plt.show()
```
