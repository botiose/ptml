```python
import pandas as pd
import numpy as np

data = pd.read_csv('data1.csv')

class1 = data[data.pclass == 1]
print(class1)
print("--------")

male_30_50 = data[(data.sex == 'male') & (data.age >= 30) & (data.age <= 50)]
print(male_30_50)
print("--------")

per_class = data.groupby('pclass')
print("EFFECTIF :")
print(per_class['pclass'].count())
print("AGE MOYEN :")
print(per_class['age'].mean())
print("SURVIE :")
print(per_class['survived'].mean())
```
