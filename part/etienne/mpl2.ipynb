```python
import pandas as pd

from matplotlib import pyplot as plt

dataset = pd.read_csv('./data1.csv')

plt.xlabel('age')
plt.ylabel('prix')

plt.scatter(dataset.age, dataset.fare)
plt.show()
```

```python
import pandas as pd

from matplotlib import pyplot as plt
from matplotlib import patches

dataset = pd.read_csv('./data1.csv')

plt.xlabel('age')
plt.ylabel('prix')

plt.scatter(dataset.age, dataset.fare, c=dataset.survived.apply(lambda d: ['yellow', 'green'][d]))

colornames = [patches.Patch(color='green', label='survived'), patches.Patch(color='yellow', label='no')]

plt.legend(handles=colornames)
plt.show()
```

```python
import pandas as pd

from matplotlib import pyplot as plt
from pandas.plotting import scatter_matrix

dataset = pd.read_csv('./data1.csv')[['fare', 'age', 'survived']].dropna(axis='index')

colors = list(map(lambda x:['red', 'green'][x], dataset.survived))

scatter_matrix(dataset[['fare', 'age']], alpha=0.5, c=colors, diagonal='kde')
plt.show()
```

```python
import pandas as pd

from matplotlib import pyplot as plt
from pandas.plotting import scatter_matrix

dataset = pd.read_csv('./data1.csv')[['fare', 'sex', 'age', 'survived']].dropna(axis='index')

dataset.sex = dataset.sex.apply(lambda x:{'male':0, 'female':1}[x])

colors = list(map(lambda x:['red', 'green'][x], dataset.survived))

scatter_matrix(dataset[['fare', 'sex', 'age']], alpha=0.5, c=colors, diagonal='hist')
plt.show()
```

```python
import pandas as pd

from matplotlib import pyplot as plt
from pandas.plotting import scatter_matrix

from random import gauss

dataset = pd.read_csv('./data1.csv')[['fare', 'sex', 'age', 'survived']].dropna(axis='index')

dataset.sex = dataset.sex.apply(lambda x:{'male':0, 'female':1}[x])

dataset.sex = dataset.sex.apply(lambda x: x + gauss(0,0.05))

colors = list(map(lambda x:['red', 'green'][x], dataset.survived))

scatter_matrix(dataset[['fare', 'sex', 'age']], alpha=0.5, c=colors, diagonal='hist')
plt.show()
```
