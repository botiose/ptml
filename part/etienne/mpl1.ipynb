```python
from matplotlib import pyplot as plt
import numpy as np
from math import exp

plt.xlabel('x')
plt.ylabel("y=exp(-lambda x)")

x = np.linspace(0,1,101) # equivalent to np.arange(0, 1, 0.01)

lambda_list = [1, 5, 10]

for my_lambda in lambda_list:
    plt.plot(x, np.vectorize(lambda x2:exp(-my_lambda * x2))(x), label=f'lambda: {my_lambda}')

plt.legend()
plt.show()

```

```python
from matplotlib import pyplot as plt
import numpy as np
from math import exp

    
fig = plt.figure()

x = np.linspace(0,1,101)

lambda_list = [2, 2, 10]

for i, my_lambda in enumerate(lambda_list):
    fig.add_subplot(2, 2, i + 1)
    plt.xlabel('x')
    plt.plot(x, np.vectorize(lambda x2:exp(-my_lambda * x2))(x), label=f'y=exp(-{my_lambda}x)')
    plt.legend()
    
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
from math import exp

plt.xlabel('x')
plt.ylabel("y=exp(-lambda x)")

x = np.linspace(0,1,101)

lambda_list = [1, 5, 10]

for my_lambda in lambda_list:
    plt.plot(x, np.vectorize(lambda x2:exp(-my_lambda * x2))(x), label=f'lambda: {my_lambda}')
    plt.annotate("y=1/e", (1 / my_lambda, exp(-1)))

plt.legend()
plt.show()
```
