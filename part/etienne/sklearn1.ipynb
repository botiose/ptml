```python
import pandas as pd
from sklearn.linear_model import LinearRegression

iris = pd.read_csv('./flowers.csv')

X = iris.PetalWidth.values.reshape(-1, 1)
Y = iris.PetalLength
lr = LinearRegression()

lr.fit(X, Y)
print(lr.coef_)
print(lr.score(X, Y))

Y = iris.SepalLength
lr.fit(X, Y)
print(lr.coef_)
print(lr.score(X,Y))
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt

iris = pd.read_csv('./flowers.csv')

X = iris.PetalWidth.values.reshape(-1, 1)
Y = iris.PetalLength
lr = LinearRegression()

lr.fit(X, Y)

plt.xlabel("Epaisseur des Pétales")
plt.ylabel("Longueur des Pétales")

plt.scatter(X, Y, c='orange')

test = np.arange(0, 2.5, 0.01).reshape(-1, 1)
plt.plot(test, lr.predict(test))

plt.show()
```
