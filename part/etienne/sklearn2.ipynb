```python
import pandas as pd
from sklearn.decomposition import PCA

df = pd.read_csv('./data2.csv', sep=';')

data = df[['adm', 'dip', 'mil']].dropna()

acp = PCA()

acp.fit_transform(data)

print(acp.explained_variance_)
print(acp.components_)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt

df = pd.read_csv('./data2.csv', sep=';')

data = df[['adm', 'dip', 'mil']].dropna()

acp = PCA()

rot = acp.fit_transform(data)

# print(rot, len(rot))
# print(rot[:,0], len(rot))
plt.scatter(rot[:,0], rot[:,1], alpha=0.5)
cf = acp.components_

plt.xlabel(f"{cf[0][0]:.2}adm+{cf[0][1]:.2}dip+{cf[0][2]:.2}mil")
plt.ylabel(f"{cf[1][0]:.2}adm+{cf[1][1]:.2}dip+{cf[1][2]:.2}mil")


plt.show()
```
